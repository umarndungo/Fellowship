# Calculate Efficiency 

`calculate_efficiency(actualOutput, targetOutput)`

This is a function that calculates the efficiency percentage of an operational unit, for example:
- A store branch
- A depot
- A manufacturing site

by dividing the `actualOutput` of the site by the `targetOutput` and multiplying by 100 to get a percentage.

```python
def calculate_efficiency(actualOutput, targetOutput):
    '''Calculates efficiency percentage
    
    Args:
        actualOutput: the output of a system
        targetOutput: the expected output from the system

    Returns:
        efficiencyPercentage: the efficiency percentage of the actualOutput relative to the targetOutput
    '''

    return (actualOutput / targetOutput) * 100
```

# Finding the Operational Status 

Using the efficiency percentage we need to give a status score for the operational site. 

`get_operational_status(efficiency)`
This is a function that assigns a status based on the efficiency percentage of an operational unit, 

The possible status values are:
- "Normal"   - if the efficiency is >= 90%
- "Warning"  - if the efficiency is >= 70% but < 90%
- "Critical" - if the efficiency is < 70%

by dividing the `actualOutput` of the site by the `targetOutput` and multiplying by 100 to get a percentage.

```python
def get_operational_status(efficiency):
    """Classifies the efficiency status of the system.

    Args:
        efficiency (float): The measured efficiency percentage of the system.

    Returns:
        str: The status classification based on the efficiency. 
            Possible values are "Critical", "Warning", or "Normal".
    """
    status = ""

    if efficiency >= 90:
        return "Normal"
    elif efficiency >= 70 and efficiency < 90:
        return "Warning"
    else:
        return "Critical"

```

# Data

`operationsSites` 

This is a dictionary representing 3 different operational sites, in our case, three store branches across Kirinyaga county in Kenya.

```python
operationsSites = {
    "mwea": {
        "name": "mwea branch",
        "actualOutPut": 4200,
        "targetOutput": 30000
    },
    "kianyaga": {
        "name": "kianyaga branch",
        "actualOutPut": 14200,
        "targetOutput": 30000
    },
    "sagana": {
        "name": "sagana branch",
        "actualOutPut": 24200,
        "targetOutput": 30000
    },
}
```

```python
results = []

for operationalSite in operationsSites.values():
    siteName = operationalSite["name"]
    # calculate efficiency percentage of the operational site
    efficiency = calculate_efficiency(operationalSite["actualOutPut"], operationalSite["targetOutput"])

    # determine the status of the site based on its efficiency percentage
    status = get_operational_status(efficiency)

    # compile the results and store in a list format
    siteScore = "The " + siteName + f" branch's status is {status}, operating at an efficiency of {efficiency}%"
    results.append(siteScore)


```
